```python
import json
import pandas as pd
```

```python
from geosky import geo_plug

geo_plug.all_CountryNames()

geo_plug.all_Country_StateNames()

dict = geo_plug.all_State_CityNames('all')# name == 'all' or stae name
#geo_plug.all_State_CityNames('Odisha')
```

```python
f = open('allCities2.json','w')
f.write(dict)
f.close()

f = open('allCities2.json')
#Returns Json object as
# a dictionary
data = json.load(f)
```

```python
list(data[7].values())[0]
```

```python
df = pd.DataFrame(columns = ['zone' ,  'city'])
for i in range(len(data)):
    print(data[i], i)
    for j in range(len(list(data[i].values())[0])):
        temp_df = pd.DataFrame({'zone' : list(data[i].keys()) , 'city':list(data[i].values())[0][j]})
        df = pd.concat([df , temp_df])
```
